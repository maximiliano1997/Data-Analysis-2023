# 3- Indexes: how to Set, Reset, and use Indexes

```python
people = {
    "first": ["Corey", "Jane", "Jhon"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchager@gmail.com", "loco@gamil.com", "lineal@gmail.com"],
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.set_index('email', inplace=True)
```

```python
df.loc['lineal@gmail.com', 'first']
```

```python
df.index
```

```python
df.reset_index(inplace=True)
df
```

### Working with Stackoverflow Data

```python
dfs.head()
```

```python
high_salary = dfs['CompTotal'] > 200000
```

```python
dfs[high_salary]
```

```python
dfs.loc[high_salary, ['Country','CompTotal','CompFreq','Currency', 'LanguageHaveWorkedWith']]
```

```python
countries = ['Argentina','Uruguay','Chile']
search = dfs['Country'].isin(countries)
```

```python
dfs[search]
```

```python
dfs.loc[search, ['Country','CompTotal']]
```

```python
containFilter = dfs['LanguageHaveWorkedWith'].str.contains('Python', na=False)
dfs[containFilter]
```

```python
containFilter
```

```python
dfs.loc[containFilter, ['CompTotal', 'LanguageHaveWorkedWith']]
```
