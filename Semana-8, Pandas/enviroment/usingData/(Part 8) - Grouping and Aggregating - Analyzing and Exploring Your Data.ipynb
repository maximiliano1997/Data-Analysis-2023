```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count':'cantidadRespondieron', 'LanguageHaveWorkedWith':'LosqueSabenPython'}, inplace=True)
python_df
```

```python
python_df['PerKnowPython'] = (
    python_df['LosqueSabenPython'] / python_df['cantidadRespondieron']) * 100
python_df
```

```python
python_df.sort_values(by='PerKnowPython', ascending=False, inplace=True)
```
