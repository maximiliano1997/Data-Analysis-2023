```python
country_respondents = dfs['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count':'cantidadRespondieron', 'LanguageHaveWorkedWith':'LosQueSabenPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PerKnowsPython'] = (python_df['LosQueSabenPython'] / python_df['cantidadRespondieron']) * 100
python_df
```

```python
python_df.sort_values(by='PerKnowsPython', ascending=False, inplace=True)
```
