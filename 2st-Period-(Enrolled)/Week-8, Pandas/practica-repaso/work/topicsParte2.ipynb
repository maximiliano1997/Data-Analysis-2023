Filtering: Using Conditionals to Filter Rows and Columns

```python
df.head()
```

```python
filt = (df['name'] == 'Dune') & (df['year'] == 2020)
```

```python
df[filt]
```

```python
df.loc[-filt]
```

```python
high_id = (df['id'] >= 25) & (df['id'] <= 30)
```

```python
df[high_id]
```

```python
df.loc[high_id, ['name', 'year']]
```

```python
args = ['Red Mars', 'Kindred', 'Dark Matter']
search = (df['name'].isin(args))
```

```python
df.loc[search, ['name', 'year']]
```

```python
containFilter = df['name'].str.contains('Dark', na=False)
containFilter
```

```python
df.loc[containFilter]
```

```python
df.loc[containFilter, 'name']
```
