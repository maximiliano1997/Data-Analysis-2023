```python
authors_group = df.groupby(df['author'])
authors_group
```

```python
authors_group.get_group('Isaac Asimov')
```

```python
filt = df['author'] == 'H.G. Wells'
df.loc[filt]['name'].value_counts()
```

```python
authors_group['author'].value_counts().loc['H.G. Wells']
```

```python
authors_group['id'].median().loc['H.G. Wells']
```

```python
authors_group['id'].agg(['median', 'mean']).loc['Isaac Asimov']
```

```python
authors_group['name'].contains('Python').sum() # ESto no funciona
```

```python
authors_group['name'].apply(lambda x: x.str.contains('dark').sum())
```

```python
groupOne = authors_group['name'].apply(lambda x: x.str.contains('dark').sum())
groupOne
```

```python
groupTwo = df['name'].value_counts()
groupTwo
```

```python
new_df = pd.concat([groupOne, groupTwo], axis='columns', sort=False)
new_df
```

```python
new_df.rename(columns={'name':'nombres', 'count':'num'}, inplace=True)
new_df
```

```python
new_df['Percentcalculacion'] = (new_df['nombres'] / new_df['num']) * 100
new_df
```
