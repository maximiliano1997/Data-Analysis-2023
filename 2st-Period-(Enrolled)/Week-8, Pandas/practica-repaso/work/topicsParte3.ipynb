1) Updating Rows and Columns (Modifying Data Within DataFrames)

```python
df = pd.read_json('./data.json')
df.head()
```

```python
df.columns
```

```python
df.columns = ['id','movie_name', 'author_name', 'year']
```

```python
df.head()
```

```python
df.columns = [x.upper() for x in df.columns]
df.tail()
```

```python
df.columns = df.columns.str.replace('_', '.')
df.head()
```

```python
df['AUTHOR.NAME']
```

```python
df.rename(columns={'ID': 'id','MOVIE.NAME': 'name', 'AUTHOR.NAME':'author', 'YEAR': 'year'}, inplace=True)
```

```python
df.loc[2, ['name','author','year']] = ['Mandingo', 'Imanol','2020'] 
df
```

```python
df.loc[5] = [6, 'XXX', 'Kendra Lust', '2020']
df
```

```python
filt = (df['name'] == 'XXX')
df.loc[filt, 'author'] = ['Vin Diesel']
df.head(10)
```

```python
df['name'] = df['name'].str.lower()
df.head(10)
```

 Methods para trabajar apply, map, applymap & replace

```python
## Apply Method
df['name'].apply(len)
```

```python
def update_name(name):
    return name.upper()

df['author'] = df['author'].apply(update_name)
df
```

```python
df['author'] = df['author'].apply(lambda x: x.lower()) # Esta es una func Lambda
df
```

```python
df.apply(len)
```

```python
df.astype(str).apply(pd.Series.min)
```

Applymap Method

```python
df.astype(str).applymap(len)
```

```python
df.astype(str).applymap(str.upper)
```

Map Method

```python
df['name'].map({'XXX':'xxx', 'mandingo': 'Mandingo'})
df.head(10)
```

```python
df['name'] = df['name'].replace({'xxx':'XXX', 'mandingo':'Mandingo'})
df.head(6)
```

2) Add or REmove Rows and Columns From DataFrames

```python
## This is how we add columns to our DataFrame
df['name'] + " " + df['author']
```

```python
df['Full_Name'] = df['name'] + "|" + df['author'] # nueva columna creada
df.head(6)
```

```python
df.drop(columns=['name', 'author'], inplace=True) # Borra esas 2 columnas
df.head(6)
```

```python
df[['name', 'author']] = df['Full_Name'].str.split("|", expand=True)
df.head(10)
```

```python
del df['Full_Name']
df.head()
```

```python
df = pd.concat([df, pd.DataFrame([{'name':'Miraclersss'}])])
df.tail()
```

```python
df.drop(index=(2), inplace=True) # Remueve un unico row
df.head()
```

```python
df.drop(index=[1,5], inplace=True) # remueve mas de 1 rows
df.head()
```

```python
df.drop(index=range(6, 8), inplace=True) # Remueve un grupo de rows (desde, hasta)
df.head(10)
```

```python
# Remover Rows con un condicional
filtro = df['name'] == 'snow crash'
df.drop(index=df[filtro].index, inplace=True)
df.head(10)
```
